# Wrangle MBTA Days with Games #

## Setup ##

```python
import sys
import datetime
import numpy as np
import pandas as pd
```

## Import Data ##

```python
original = pd.read_csv("../../../data/mbta_daily_noholidays.csv", low_memory=False)
original.head()
```

```python
games = pd.read_csv("../../../data/game_days.csv", low_memory=False)
games.head()
```

```python
subgames = games[['date', 'bruins_game', 'celtics_game', 'sox_game']]

original['service_day'] = pd.to_datetime(original['service_day'])
subgames['date'] = pd.to_datetime(subgames['date'])

out = pd.merge(original, subgames, how='left', left_on='service_day', right_on='date')


print(len(original))
print(len(out))

print(len(subgames['date']))
print(len(subgames['date'].unique()))

out.head()
```

```python
out.to_csv("../../../data/mbta_daily_noholidays_games.csv", index=False)
```
