# Wrangle MBTA Days with Groupings from Similar Station Analysis #

## Setup ##

```python
import sys
import datetime
import numpy as np
import pandas as pd
```

## Import Data ##

```python
original = pd.read_csv("../../../data/mbta_daily_weather_games_noholidays_weekdays.csv", low_memory=False)
original.head()
```

```python
station_groupings = pd.read_csv("../../../data/stations_clean.csv", low_memory=False)
print(station_groupings.columns)
station_groupings.head()
```

```python
substation_groupings = station_groupings[['stationid', 'grouping']].copy()

out = pd.merge(original, substation_groupings, how='left', left_on='locationid', right_on='stationid')

#out.drop('date', axis=1, inplace=True)
#out['bruins_game'] = out['bruins_game'].fillna(0)
#out['celtics_game'] = out['celtics_game'].fillna(0)
#out['sox_game'] = out['sox_game'].fillna(0)


print(len(original))
print(len(out))

#print(np.unique(out['grouping']))
print(out['grouping'])

out.head()
```

```python
out.to_csv("../../../data/mbta_daily_weather_games_grouping_noholidays_weekdays.csv", index=False)
```
