## Setup

```python
# Libraries.
import matplotlib, matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns

# Setup.
% matplotlib inline

# Paths.
path_data = '../../../data/gatecount_%d/'
```

## Loading &amp; Cleaning

```python
# Get all filenames.
years = [2014,2015]
gc_files = []
for year in years:
    year_files = [(path_data % year + f) for f in os.listdir(path_data % year) if f[-4:] == '.csv']
    gc_files = gc_files + year_files
    
# Load all files as DFs and concatenate.
gate_count = [pd.read_csv(path) for path in gc_files]
gate_count = pd.concat(gate_count)
gate_count.reset_index(inplace=True,drop=True)
```

### Clean Up Dates

Dates are currently strings and integers. We'll want to use formal date/time representations for them.

```python
# Convert main date column.
gate_count.servicedate = pd.to_datetime(gate_count.servicedate)
# Convert service time to min 0. Nevermind: it's unclear how we should do this.
# gate_count.servicetime = gate_count.servicetime-300
# Convert service time to minutes.
gate_count['servicetimeminutes'] = np.floor((gate_count.servicetime-300) / 100)*60 + np.mod(gate_count.servicetime,100)

# Populate our dedicated datetime column with converted servicedate values.
gate_count['servicedatetime'] = gate_count.servicedate
# Add the timedeltas.
gate_count['servicedatetime'] += np.array(gate_count.servicetimeminutes, dtype='timedelta64[m]')
```

```python
# represent time as hour.fraction_of_hour to make it on a base-100 system
# instead of a base-60-minute system. Will make it easier to display on x-axis without scaling.
gate_count['servicetime_frac']=np.floor(gate_count['servicetime']/100)+(gate_count['servicetime']%100)/60

temp=[]
temp2=[]
#this processing runs for a long time, couldn't figure out a way to utilize 'map()' to speed up the process
for i in range(len(gate_count)):
    temp.append(gate_count.servicedate[i].weekday())
    temp2.append(gate_count['servicedate'][i].date().month)


# set a series where Monday corresponds to 0, Tues to 1,..., Sunday to 6
gate_count['weekday']=temp
gate_count['month']=temp2
```

## Distributions

Charts below show distributions at various levels of aggregation: non-aggregated (effectively by location and date/time), by date, and by location/date.

```python
# Store several convenient distributions.

# Date.
gate_count_by_date = gate_count[['servicedate','entries','exits']].groupby('servicedate').agg(np.sum)

# Location-date.
gate_count_by_location_date = gate_count[['locationid','servicedate','entries','exits']].groupby(['locationid','servicedate']).agg(np.sum)

# Time.
gate_count_by_time = gate_count[['servicetimeminutes','entries','exits']].groupby('servicetimeminutes').agg(np.sum)
```

```python
# Plot entries, exits by time. Very rough plot (only using minutes of day).
plt.figure(figsize=(12,8))
plt.plot(gate_count_by_time.index, gate_count_by_time.entries,      label='Entries')
plt.plot(gate_count_by_time.index, gate_count_by_time.exits,   'r', label='Exits')
plt.title('Entries/Exits by Date')
plt.xlabel('Time (Minutes of Day)')
plt.legend()
plt.show()
```

Overlap between entries and exits looks pretty intuitive: in the aggregate, people enter and then seem to exit somewhat later. We lose some exits due to gates staying open for >1 person and people getting off in ways that aren't tracked. But visually, it looks like we "lose" more exits later in the day. I wonder if that happens as people board through a subway entrance but exit (untracked) off of the Green Line. Maybe farther-out stops have more exits without gates.

Also, it looks like orignial `servicetime` values < 2400 are all accurate and that values ≥ 2400 actually represent times after midnight. A morning peak at ~300 (as shown) is 5am; adding three hours would put the morning peak at 8am (much more reasonable). So to clean up times, we should *not* simply subtract 300.

###Visualizing Trends By Day of Week

The goal of this section is to explore the 15-minute aggregated MBTA gate entrances for each day of the week.

```python
# groupby both day of week and time_fraction
gate_count_by_dow_time = gate_count[['weekday','servicetime_frac','entries']].groupby(['weekday','servicetime_frac'])

dow_time_mean=gate_count_by_dow_time.agg(np.mean)
dow_time_std=gate_count_by_dow_time.agg(np.std)
```

```python
# When computing confidence intervals, the std dev for some time intervals is so large that the interval
# encompasses a negative amount of entries. This function just limits the entries to be >=0
def stay_pos(x):
    if(x<0):
        return 0;
    return x;

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(12, 15), 
                         tight_layout=True)
#print axes.ravel()
days=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
for ax, day in zip(axes.ravel(), range(7)):

    ax.plot( dow_time_mean.loc[day].index, dow_time_mean.loc[day]['entries'], color='r', )
    
    upper=(dow_time_mean.loc[day]['entries'] + dow_time_std.loc[day]['entries']).values

    lower=(dow_time_mean.loc[day]['entries'] - dow_time_std.loc[day]['entries']).values
    lower=map(stay_pos,lower)
    
    ax.fill_between(dow_time_mean.loc[day].index, upper, lower, color='b', alpha=.2)
    
    labels=['','5:00 am','10:00 am','3:00 pm','8:00 pm','1:00 am','']
    
    ax.xaxis.set_ticklabels(labels,rotation=25)
    ax.set_ylabel('Mean Entries')
    ax.set_xlabel('Hour of day')
    ax.set_title(days[day])
```

###Comments:
Computing the mean entries for each time interval for each day does an adequate job of displaying the volume across all MBTA stations over the course of the day. It's clear that some time intervals exhibit much more volatility than others, as indicated by the blue shading showing one standard deviation from the mean. However, it is unclear how much of the volatility is due to consumer demand and how much volatility is due to the fact that we're aggregating over both the busiest and slowest stations. A next logical step would be to select the top 10 busiest stations and the top 10 slowest stations to see how their volatilities compare for each day and time interval. This analysis can also be carried out on individual stations while controling for other effects like precipitation, a baseball game, etc.

```python

# function to create the same day-by-day visuals as above, but only aggregated for one station of interest

def plot_station(stat_id):
    
    station_records=gate_count[gate_count['locationid']==stat_id];
    
    # groupby both day of week and time_fraction
    station_grouping = station_records[['weekday','servicetime_frac','entries']].groupby(['weekday','servicetime_frac'])

    dow_time_mean=station_grouping.agg(np.mean)
    dow_time_std=station_grouping.agg(np.std)

    fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(12, 15), 
                             tight_layout=True)
    #print axes.ravel()
    days=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
    for ax, day in zip(axes.ravel(), range(7)):

        ax.plot( dow_time_mean.loc[day].index, dow_time_mean.loc[day]['entries'] ,color='r')
        ax.scatter(dow_time_mean.loc[day].index, dow_time_mean.loc[day]['entries'], alpha=.04, lw=0)

        upper=(dow_time_mean.loc[day]['entries'] + dow_time_std.loc[day]['entries']).values

        lower=(dow_time_mean.loc[day]['entries'] - dow_time_std.loc[day]['entries']).values
        lower=map(stay_pos,lower)

        ax.fill_between(dow_time_mean.loc[day].index, upper, lower, color='b', alpha=.2)

        labels=['','5:00 am','10:00 am','3:00 pm','8:00 pm','1:00 am','']


        ax.xaxis.set_ticklabels(labels,rotation=25)
        ax.set_ylabel('Mean Entries')
        ax.set_xlabel('Hour of day')
        ax.set_title(days[day]+", Station id: "+str(stat_id))
```

```python
def plot_am_stats(stat_id, cold_months=False):

    ## Going to look at accumulation of total riders from opening until 9:30
    station_records=gate_count[gate_count['locationid']==stat_id];
    
    #check if parameter passed to compare to cold months
    if(cold_months==True):
        station_records=station_records.loc[station_records['cold']]



    #only get Mon-Fri records
    station_records=station_records[station_records['weekday']<5]

    #only look at times before 9:30, want to look at entries before end of rush hour
    station_records=station_records[(station_records['servicetime_frac']<=9.5) & (station_records['servicetime_frac']>=6.5) ]

    # groupby both day of week and time_fraction
    station_grouping = station_records[['weekday','servicetime_frac','entries','cold']].groupby(['weekday','servicetime_frac'])
    dow_time_mean=station_grouping.agg(np.mean)

    fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8,8), 
                             tight_layout=True)
    
    #place holder for dataframe to be returned
    current=[]
    
    for day in range(5):
        current=dow_time_mean.loc[day]
        total_for_rush=current['entries'].sum()

        cumulatives=[]
        sums=[]
        #rate of change array
        roc=[]
        for i in range(len(current)):
            #calculate cumulative sum up to this point in rush hour
            sum_=current.iloc[0:i+1].entries.sum()
            sums.append(sum_)
            cumulatives.append(sum_/total_for_rush)

            if(i==0):
                roc.append(0)
            else:
                roc.append((current.iloc[i].entries-current.iloc[i-1].entries)/current.iloc[i-1].entries)

        # cumulative sum of all entries that morning prior to interval t
        current['sums']=sums
        # same as cumulative sum, expressed as precentage of total entries prior 6:30-9:30, much like CDF
        current['cdf']=cumulatives
        # rate of change at time t, roc[t]=(entries[t]-entries[t-1])/entries[t-1]
        current['roc']=roc

        days=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
        ## Plotting time        
        ax[0].plot( current.index, current['roc'],label=days[day])

        labels=['','5:00 am','10:00 am','3:00 pm','8:00 pm','1:00 am','']

        if(cold_months):
            title=". Cold months "
        else:
            title=". Not Cold Months "
            
        #ax.xaxis.set_ticklabels(labels,rotation=25)
        ax[0].set_ylabel('Rate of Change Entries')
        ax[0].set_xlabel('Hour of day')     
        ax[0].set_title("Rate of change, Station id: "+str(stat_id)+title)
        ax[0].legend(loc="upper right")
        
        ## Plotting time        
        ax[1].plot( current.index, current['cdf'],label=days[day])

        #ax.xaxis.set_ticklabels(labels,rotation=25)
        ax[1].set_ylabel('CDF entries')
        ax[1].set_xlabel('Hour of day')
        ax[1].set_title("CDF of Entries, Station id: "+str(stat_id)+title)
        ax[1].legend(loc="upper left")
            
    return current
        
c=plot_am_stats(1087,True)
```
