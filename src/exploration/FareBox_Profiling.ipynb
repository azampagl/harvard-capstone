# FareBox: Profiling

This notebook profiles the FareBox files.

## Setup

```python
# Libraries.
import matplotlib, matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns

# Setup.
% matplotlib inline

# Paths.
path_data = '../../../data/farebox_%d/'
```

## Loading &amp; Cleaning

```python
# Get all filenames.
years = [2014,2015]
fb_files = []
for year in years:
    year_files = [(path_data % year + f) for f in os.listdir(path_data % year) if f[-4:] == '.csv']
    fb_files = fb_files + year_files
    
# Load all files as DFs and concatenate.
fare_box = [pd.read_csv(path) for path in fb_files]
fare_box = pd.concat(fare_box)
fare_box.reset_index(inplace=True,drop=True)
```

## Structure

```python
# Show types.
print 'Types:'
print fare_box.dtypes
# Show nulls.
print '\nNulls:'
print np.sum(fare_box.isnull())
# Show min/max service times.
print '\nMin service time: %d' % fare_box.servicetime.min()
print 'Max service time: %d' % fare_box.servicetime.max()
# Num records.
print '\nNumber of records: %d' % len(fare_box)
```

### Clean Up Dates

As with GateCount, dates and times need to be processed. For the time being, process times like we did speculatively for GateCount.

```python
# Convert main date column.
fare_box.servicedate = pd.to_datetime(fare_box.servicedate)

# Add a calendar date column.
fare_box['calendardate'] = fare_box['servicedate']
# Increment dates after midnight.
fare_box.ix[fare_box['servicetime'] >= 2400,'calendardate'] = fare_box.ix[fare_box['servicetime'] >= 2400,'calendardate']+np.timedelta64(1,'D')

# Add a calendar time minutes column and put on proper scale.
fare_box['calendartimeminutes'] = (np.floor((fare_box.servicetime) / 100)*60 + np.mod(fare_box.servicetime,100)) % (24.*60.)
# Convert to hours as well.
fare_box['calendartimehours'] = fare_box['calendartimeminutes'].astype(np.float) / 60

# Populate our dedicated datetime column with converted servicedate values.
fare_box['calendardatetime'] = fare_box.calendardate
# Add the timedeltas.
fare_box['calendardatetime'] += np.array(fare_box.calendartimeminutes, dtype='timedelta64[m]')
```

## Summary Statistics

### Distributions

```python
# Store several convenient distributions.

# Date.
fare_box_by_date = fare_box[['calendardate','charlie','cash']].groupby('calendardate').agg(np.sum)

# Signcode-date.
fare_box_by_location_date = fare_box[['signcodeid','calendardate','charlie','cash']].groupby(['signcodeid','calendardate']).agg(np.sum)

# Time.
fare_box_by_time = fare_box[['calendartimehours','charlie','cash']].groupby('calendartimehours').agg(np.sum)
```

```python
# Plot distribution of entries.
plt.figure(figsize=(10,6))
plt.hist(fare_box.charlie+fare_box.cash, bins=40)
plt.vlines((fare_box.charlie+fare_box.cash).mean(),0,plt.ylim()[1],colors='r')
plt.title('Entries by Signcode-Date-Time')
plt.xlabel('Entries') ; plt.ylabel('Frequency')
plt.show()
```

```python
# Plot distribution of entries by date.
plt.figure(figsize=(10,6))
plt.hist(fare_box_by_date.charlie+fare_box_by_date.cash, bins=40)
plt.vlines((fare_box_by_date.charlie+fare_box_by_date.cash).mean(),0,plt.ylim()[1],colors='r')
plt.title('Entries by Date')
plt.xlabel('Entries') ; plt.ylabel('Frequency')
plt.show()
```

### Time Trends

```python
# Plot entries by date.
plt.figure(figsize=(12,8))
plt.plot(fare_box_by_date.index,fare_box_by_date.charlie+fare_box_by_date.cash)
plt.title('Entries by Date')
plt.xlabel('Date') ; plt.ylabel('Entries')
plt.show()
```

```python
# Plot entries, exits by time. Very rough plot (only using minutes of day).
plt.figure(figsize=(12,8))
plt.plot(fare_box_by_time.index, fare_box_by_time.charlie+fare_box_by_time.cash)
plt.title('Entries by Time')
plt.xlabel('Time (Hour of Day)')
plt.xticks(np.arange(0,25,2))
plt.show()
```
