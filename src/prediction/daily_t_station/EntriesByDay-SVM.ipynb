```python
import sys
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib, matplotlib.pyplot as plt

from sklearn import cross_validation
from sklearn import grid_search

from sklearn import linear_model
from sklearn import svm

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# Setup.
% matplotlib inline
```

```python
data = pd.read_csv("mbta_daily_checkpoint03.csv", low_memory=False)
data.head()
```

```python
print("Original: " + str(len(data)))

data = data[(pd.DatetimeIndex(data['service_day']).weekday != 5) & (pd.DatetimeIndex(data['service_day']).weekday != 6)]

print("Remaining: " + str(len(data)))
data.head()
```

```python
#
cols = list(data.columns)
y_cols = ['entries']
#x_cols = ['day_of_week_0', 'day_of_week_1', 'day_of_week_2', 'day_of_week_3', 'day_of_week_4', 'entries_weeks_ago_1', 'entries_weeks_ago_2', 'entries_weeks_ago_3']
x_cols = ['entries_weeks_ago_1', 'entries_weeks_ago_2']

#
y_indices = [0] + list(np.where([col in y_cols for col in cols])[0] + 1)
x_indices = [0] + list(np.where([col in x_cols for col in cols])[0] + 1)

#
station = data[data['locationid'] == 1002]

#
print("Before Filter: " + str(len(station)))
for col in x_cols:
    station = station[pd.notnull(station[col])]
print("After Filter: " + str(len(station)))

#
matrix = station.reset_index().values

#
y = matrix[:,y_indices]
x = matrix[:,x_indices]

#
x_train, x_test, y_train, y_test = cross_validation.train_test_split(x, y, test_size=0.2, random_state=0)

parameters = {'kernel': ('linear', 'rbf'), 'C': range(1, 2)}
predictor = svm.SVR()

#predictor = linear_model.LinearRegression()
#parameters = {}

#
x_train_fit = np.matrix(x_train[:,1:], dtype=np.uint16)
y_train_fit = np.array([v[0] for v in y_train[:,1:]], dtype=np.uint16)

x_test_fit = x_test[:,1:]
y_test_fit = [v[0] for v in y_test[:,1:]]

#
clf = grid_search.GridSearchCV(predictor, parameters, scoring='mean_squared_error', cv=10)
clf.fit(x_train_fit, y_train_fit)

#
model = clf.best_estimator_
print(model)

#
y_pred_fit = model.predict(x_test_fit)

#
pdiffs = np.abs(y_pred_fit - y_test_fit) * 100 / y_test_fit

#
pdiffs = pdiffs[pdiffs < float('+inf')]
print(np.mean(pdiffs))
# Elimi
#outlier_indices = y_test[pdiffs > 20][:,0]

#print(data.iloc[outlier_indices])
#print(outlier_indices)
#print(y_test[pdiffs > 20])
#print(y_pred_fit[pdiffs > 20])

#print(pdiffs)
```
