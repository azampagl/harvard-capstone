## Setup

```python
import sys
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sys.path.append("../../")
import features
import features.date
import features.entries
import features.station
import features.weather
```

```python
data_raw = pd.read_csv("../../../../data/mbta.csv")
data_raw.head()
```

```python
data_grouped = data_raw.groupby(['locationid', 'service_day'], as_index=False).agg({'entries': np.sum})

print("Original: " + str(len(data_raw)))
print("Remaining: " + str(len(data_grouped)))
data_grouped.head()
```

```python
data = pd.merge(
    data_grouped,
    data_raw.drop_duplicates(['locationid', 'service_day']),
    how='left',
    on=['locationid', 'service_day'])

# The following columns we will likely not use in our research.
data.drop(['entries_y', 'exits', 'wind_speed', 'vis_min', 'vis_max', 'vis_mean'], axis=1, inplace=True)

data.rename(columns={'entries_x': 'entries'}, inplace=True)
data.head()
```

```python
print("Original: " + str(len(data)))

data = data[(pd.DatetimeIndex(data['service_day']).weekday != 5) & (pd.DatetimeIndex(data['service_day']).weekday != 6)]

print("Remaining: " + str(len(data)))
data.head()
```

```python
data = features.date.init(data)
data = features.date.add_day_of_week_binary(data)
data = features.date.add_month_binary(data)
data.head()
```

```python
data.to_csv("mbta_daily_checkpoint01.csv", index=False)
print("done")
```

```python
data = features.entries.init(data)
data = features.entries.add_previous_week(data, weeks_ago=1)
data = features.entries.add_previous_week(data, weeks_ago=2)
data = features.entries.add_previous_week(data, weeks_ago=3)
data.head()
```

```python
data.to_csv("mbta_daily_checkpoint02.csv", index=False)
print("done")
```
